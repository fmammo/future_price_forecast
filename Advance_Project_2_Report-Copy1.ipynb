<h2 style ="font-family:Bell MT;font-size:40px:text-align:center">1.Introduction<h2/>

<p style="font-family:Bell MT;font-size:18px;text-align:Justify">A data which is collected based on time is called time series. Now a days, based on my observation price forecasting is a very important aspect for so many businesses. By considering the number of time dependent variables it has, time series forecasting is divided in to two: univariate and multivariate. Univariate is a series with a single time-dependent variable whereas multivariate is a series that has more than one time-dependent variable. This project deals with a multivariate time series forecasting mainly for future price. The project is one part of a research on the price establishment at the futures market and cocoa processors in Germany. Cocoa is a vital agricultural export outcome for West African countries which work for the European market. There is several sentiment news which has a relation with cocoa future price, for instance: Social Fear, weather damage, social supply Vs demand forecast, social stress, emotion Vs fact, price forecast news and so on.</p>
<p style="font-family:Bell MT;font-size:18px;text-align:Justify">This cocoa future price forecast project is mainly based on weather damage sentiment news and price forecast sentiment news with previous close price values. Weather damage and price forecast sentiment news are divided in two three parts individually. Thomson and Reuters (TR) sentiment is a tool that gives a certain number to the news, it might be negative or postive. So, the two-sentiment news has a value given by TR tool. Generally, there are 4808 observation by 7-time dependent variables. Through all the report of this project EDA which stands for ‘exploratory data analysis’ for all content of the dataset and other time series analysis are performed. To forecast future price of cocoa a popular multivariate time series statistical model called vector auto regression (VAR) is used. All the details about this project is briefly explained on the rest of the report.</p>

<h3 style="font-family:Bell MT">3.3.5PCA-Plot<h3/>

<p style="font-family:Bell MT;font-size:18px;text-align:Justify"> "PCA" stands for 'principal component analysis' and it is one way of identifying patterns in data and expressing it to highlight their similarities and differences. And, we can say that it is one way of dimensionality reduction.
Let me explain what pca does in a very short way: first it finds patterns in the data and it compresses the data through minimizing the number of dimensions, with out much loss of information. It calculates the projection of the original data into the same number of fewer dimensions.
Mainly we used PCA for two purposes one for data visualization and the other is to speed up during fitting of a machine learnning algorithm. Here in my case, I have used PCA for data visualization purpose. I have compressed weather damage news data together as one and price forecast news as one.
Under, the first plot shows the pca plot of ‘weather damage’ news and ‘price forecast’ whereas the second plot shows the relation between ‘close price’ and ‘price forecast’ news. Then again after that the plot shows ‘close price’ and ‘weather damage’ news. Lastly, I have tried to check or show all the time dependent variables relationship in 3D plot.
<p/>

<h3 style = "font-family:Bell MT;font-size:40px:text-align:center">5.Transformation(change-the-non-stationary-to-stationary)<h3/>

<p style="font-family:Bell MT;font-size:18px;text-align:Justify">When we change the time series stationary, we are making the data to have a constant mean, variance and covariance with time. There are a couple of methods to change non-stationary to stationary, For example: differencing, seasonal differencing and log transform. In this case I am going to use differencing method. Because most statistical models require the series to be stationary to do effective prediction Let's apply first differencing on the trainning data set and check the stationarity of the 4 series by using of ‘ADF ‘ (Augmented Dickey Fuller) Test.Below two table shows sample five dataset and their statistical numeric description after changing the series to stationary using of differencing method<p/>

<h3 style = "font-family:Bell MT;font-size:40px:text-align:center">Stationarity Check<h3/>

```python
def adfuller_test(series, signif=0.05, name='', verbose=False):
    r = adfuller(series, autolag='AIC')
    output={'test_statistic':round(r[0], 4), 'pvalue':round(r[1], 4), 'n_lags':round(r[2], 4), 'n_obs':r[3]}
    p_value = output['pvalue']
    def adjust(val, length=6): return str(val).ljust(length)
    
    print(f'    Augumented Dickey-Fuller Test on "{name}"', "\n  ", '-'*47)
    print(f' Null Hypothesis: Data has unit root. Non-Stationary.')
    print(f' Significance Level   ={signif}')
    print(f' Test Statistic       ={output["test_statistic"]}')
    print(f' No. Lags Chosen      ={output["n_lags"]}')
          
    for key,val in r[4].items():
          print(f' Critical value {adjust(key)} = {round(val, 3)}')
          
    if p_value <= signif:
          print(f" => p-value = {p_value}. Rejecting Null Hypothesis.")
          print(f" => Series is stationary.")
    else:
          print(f" => P-value = {p_value}. Weak evidence to reject the Null Hypothesis.")
          print(f" => Series is Non-Stationary.")
          
# ADF test on each column
for name, column in tras_df_merge7.iteritems():
          adfuller_test(column, name=column.name)
          print('\n')
```

<h3 style = "font-family:Bell MT;font-size:40px:text-align:center">6.Testing-causation<h3/>

<p style="font-family:Bell MT;font-size:18px;text-align:Justify">In VAR model the most important part is, if the time series in the system influence each other. Then, one can predict the series with past values of itself along with other time dependent variable or series in the system. Let’s check the relationship between the series before building the model by using of Granger's Causality Test.<p/> 

```python
from statsmodels.tsa.stattools import grangercausalitytests
maxlag=12
test = 'ssr_chi2test'
def grangers_causation_matrix(data, variables, test='ssr_chi2test', verbose=False):
    df_merge7_train = pd.DataFrame(np.zeros((len(variables), len(variables))), columns=variables, index=variables)
    for c in df_merge7_train.columns:
        for r in df_merge7_train.index:
            test_result=grangercausalitytests(data[[r, c]], maxlag=maxlag, verbose=False)
            p_values = [round(test_result[i+1][0][test][1],4) for i in range(maxlag)]
            if verbose: print(f'y = {r}, X = {c}, p values = {p_values}')
            min_p_value = np.min(p_values)
            df_merge7_train.loc[r, c] = min_p_value
    df_merge7_train.columns = [var +'_x' for var in variables]
    df_merge7_train.index = [var + '_y' for var in variables]
    return df_merge7_train
grangers_causation_matrix(df_merge7_train, variables = df_merge7_train.columns)
```

<p style="font-family:Bell MT;font-size:18px;text-align:Justify">In the above table the row is the response (y) and the columns are the predictor (x). By looking at the p-values inside the table we can determine that if one variable causes of the other variable. For instance, the 0.0 in (row 2, column 1) refers the p-value of 'mpynprfr_y' causing 'Close_price_x' and so on. Generally, by looking at the p-values we can say that most of the variables causes each other.This justifies the VAR modeling is a good approach for this multi time-series forecasting.<p/> 

<h3 style = "font-family:Bell MT;font-size:40px:text-align:center">7.Co-integration-test<h3/>

<p style="font-family:Bell MT;font-size:18px;text-align:Justify"> Co-integration test is for to stablish the presence of a statistically significant connection between the variables. We can say that there is a statistically significant relationships between the time series then the collection of series is said to be co-integrated. Let’s use Johansen co-integration test for this purpose so, I am going to perform on the original data-set.<p/>

```python
from statsmodels.tsa.vector_ar.vecm import coint_johansen

def cointegration_test(tras_df_merge7, alpha=0.05): 
    """Perform Johanson's Cointegration Test and Report Summary"""
    out = coint_johansen(tras_df_merge7,-1,5)
    d = {'0.90':0, '0.95':1, '0.99':2}
    traces = out.lr1
    cvts = out.cvt[:, d[str(1-alpha)]]
    def adjust(val, length= 6): return str(val).ljust(length)

    # Summary
    print('Name   ::  Test Stat > C(95%)    =>   Signif  \n', '--'*20)
    for col, trace, cvt in zip(tras_df_merge7.columns, traces, cvts):
        print(adjust(col), ':: ', adjust(round(trace,2), 9), ">", adjust(cvt, 8), ' =>  ' , trace > cvt)

cointegration_test(df_merge7_train)
```

<p style="font-family:Bell MT;font-size:18px;text-align:Justify"> We use Durbin-Watson Statstics to check for serial correlation of residuals (errors). We can say that there is some pattern in the time series that is left to be explained by the model, if there is any correlation left in the residuals. So, Serial correlation check is in order to ensure that the model is sufficiently able to explain the variances and patterns in the time series. In this statistic always have a value in between 0 and 4 and if the value is 2.0 That means there is no any autocorrelation. Whereas the value of 0 to > 2 indicates a postive auto correlation and a value from 2 to 4 shows a negative correlation. For the best the statistic value in the range of 1.5 to 2.5 are relatively normal. Any value outside this range will bring some concern. As it is showed below based on the above explanation, I can conclude that there is no autocorrelation exist; so, I can proceed with the forecast.<p/>
